```python
import numpy as np
import sklearn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('./pics.json')
```

```python
keys = df["data"][0].keys()
```

```python
for key in keys:
    df[key] = df["data"].map(lambda x: x[key])
    
```

```python
pixel_keys = df["pixelData"][0].keys()
for key in pixel_keys:
    df[key] = df["pixelData"].map(lambda x: x[key])
```

```python
df = df.drop("completeExif", axis=1)
df = df.drop("data", axis=1)
df = df.drop("pixelData", axis=1)
```

```python
df.head(10)
```

```python
sns.jointplot(data=df, x='creation', y='value', kind='reg', color='red')
plt.show()
```

```python
sns.jointplot(data=df, x='hue', y='value', kind='reg', color='red')
plt.show()
```
